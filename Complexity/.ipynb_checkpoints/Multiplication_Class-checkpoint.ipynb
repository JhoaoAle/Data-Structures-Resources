## School multiplication method

```c++
auto school_multiply = [](long long x, long long y) -> long long {
    long long result = 0;
    int current_position = 1;

    while (y > 0){
        int current_digit = y%10;
        printf("Y before %lld \n",y);
        y /= 10;
        printf("Y after update %lld \n",y);

        long long partial = current_digit * current_position * x;

        printf("partial result %lld \n",partial);

        current_position *= 10;
        result += partial;
    }

    return result;
};

printf("%lld", school_multiply(25600000,1200));
```
